# EALC Demo Interaction
This notebook demonstrates a sample learner interaction with the EALC (Emotion-Adaptive Learning Companion) system.

```python
# Import orchestrator
from orchestrator_agent import EALCOrchestrator

# Initialize the orchestrator
ealc = EALCOrchestrator()
```

## Example 1 — Frustrated Learner

```python
frustrated_input = {
    "learner_id": "student_123",
    "query": "I don't understand this algebra problem at all! I've been trying for an hour and nothing makes sense!",
    "subject": "mathematics",
    "difficulty_level": "intermediate",
    "learner_response": None
}

response_frustrated = ealc.process_learner_input(frustrated_input)
response_frustrated
```

## Example 2 — Confident Learner

```python
confident_input = {
    "learner_id": "student_456",
    "query": "I solved that calculus problem easily! What's next?",
    "subject": "mathematics",
    "difficulty_level": "advanced",
    "learner_response": None
}

response_confident = ealc.process_learner_input(confident_input)
response_confident
```

## Observing the Responses
- Each response includes:
  - Emotional acknowledgment
  - Personalized explanation
  - Practice questions
  - Encouragement and motivation
  - Progress tracking

```python
# Print formatted responses
for learner_response in [response_frustrated, response_confident]:
    print("---")
    print(f"Emotional Acknowledgment: {learner_response['emotional_acknowledgment']}")
    print(f"Explanation: {learner_response['explanation']}")
    print(f"Practice Questions: {learner_response['practice_questions']}")
    print(f"Encouragement: {learner_response['encouragement']}")
    print(f"Progress: {learner_response['progress']}")
```
